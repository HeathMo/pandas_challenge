```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
purchase_data_df.head()
```

```python
players_count = len(purchase_data_df["SN"].unique())
players_df = pd.DataFrame({"Total Players": [players_count]})
players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = len(purchase_data_df["Item ID"].unique())
total_revenue = purchase_data_df["Price"].sum()
purchase_number = purchase_data_df["Purchase ID"].count()
average_price = (total_revenue / purchase_number)

summary_df = pd.DataFrame({"Number of Unique Items": [item_count],
                          "Average Price": [average_price],
                          "Number of Purchases": [purchase_number],
                          "Total Revenue": [total_revenue]})
#summary_df["Average Price"].dtype
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#create new df that reduces to unique players then filter by gender
#add column of percents(6/22 notes Test Score Summary)
trim_data = purchase_data_df.drop_duplicates(subset=["SN"])
#trim_data.head()
gender_count = trim_data["Gender"].value_counts()
#print(gender_count)

#reminder players_count already created

percent_players = (gender_count/players_count)*100
#print(percent_players)

gender_percent_df = pd.DataFrame({"Total Count": gender_count,
                                 "Percentage of Players": percent_players})
gender_percent_df["Percentage of Players"] = gender_percent_df["Percentage of Players"].map("{:.2f}%".format)
gender_percent_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#reminder to use gender_count from above
purchase_by_gender = purchase_data_df.groupby("Gender")
#print(purchase_by_gender)

purchase_count = purchase_by_gender["Purchase ID"].count()
#print(purchase_count)

avg_purchase_price = purchase_by_gender["Price"].mean()
#print(avg_purchase_price)

total_purchase = purchase_by_gender["Price"].sum()
#print(total_purchase)

avg_total_person = total_purchase/gender_count
#print(avg_total_person)

purchase_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price,
                           "Total Purchase Value": total_purchase, "Avg Total Purchase per Person": avg_total_person})
purchase_df["Average Purchase Price"] = purchase_df["Average Purchase Price"].map("${:.2f}".format)
purchase_df["Total Purchase Value"] = purchase_df["Total Purchase Value"].map("${:.2f}".format)
purchase_df["Avg Total Purchase per Person"] = purchase_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_df


```

* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_group_series = pd.cut(purchase_data_df["Age"], bins, labels=group_names)
age_count = age_group_series.value_counts()
#age_count
total_count = age_group_series.count()
age_percent = (age_count / total_count)*100
#print(age_percent)

age_df = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_percent})
age_df["Percentage of Players"] = age_df["Percentage of Players"].map("{:.2f}%".format)
age_df = age_df.sort_index()
age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_by_player = purchase_data_df.groupby("SN")
#print(purchase_by_player)

#group_SN = purchase_data_df.groupby("SN")["Purchase ID"].nunique()

purchase_count_player = purchase_by_player["Purchase ID"].count()

total_purchase_player = purchase_by_player["Price"].sum()

avg_price_player = purchase_by_player["Price"].mean()
#print(avg_purchase_price)

total_player_value = (purchase_count_player * avg_price_player)

spendors_df = pd.DataFrame({"Purchase Count": purchase_count_player, "Average Purchase Price": avg_price_player,
                           "Total Purchase Value": total_player_value})
spendors_df["Average Purchase Price"] = spendors_df["Average Purchase Price"].map("${:.2f}".format)
spendors_df["Total Purchase Value"] = spendors_df["Total Purchase Value"].map("${:.2f}".format)
spendors_df = spendors_df.sort_values(by="Purchase Count", ascending=False)
spendors_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item = purchase_data_df.iloc[:,4:]
popular_item.drop('Age Summary', axis=1, inplace=True)
popular_item.head()
```

```python
item_group = popular_item.groupby(["Item ID", "Item Name"])

#purchase count
item_count = item_group["Price"].count()

#average item price
item_average_price = item_group["Price"].mean()

#total purchase value
item_total_purchase = item_group["Price"].sum()

item_prices = pd.DataFrame({"Purchase Count": item_count, "Item Price": item_average_price,
                           "Total Purchase Value": item_total_purchase})
item_prices["Item Price"] = item_prices["Item Price"].map("${:.2f}".format)
item_prices["Total Purchase Value"] = item_prices["Total Purchase Value"].map("${:.2f}".format)
item_prices = item_prices.sort_values(by="Purchase Count", ascending=False)
item_prices.head()
```
